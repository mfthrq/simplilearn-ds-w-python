```python
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
web_costumers = [123,324,563,1234,1543,1351,1235]
time_hrs = [7,8,9,10,11,12,13]
```

```python
style.use('ggplot')

#x_axis & y_axis
plt.plot(time_hrs, web_costumers, color='b', linestyle="--", linewidth="4")
plt.title('Website Traffic')
plt.xlabel("Hours")
plt.ylabel("Number of costumers")

plt.show()
```

```python
# multiple plots

web_monday = [123,324,563,1234,1543,1351,1235]
web_tuesday = [98,92,100,103,1030,400,200]
web_wednesday = [103, 204, 302, 500, 150, 500, 1000]
time_hrs = [7,8,9,10,11,12,13]
```

```python
style.use('ggplot')

plt.plot(time_hrs, web_monday, 'r', label="monday")
plt.plot(time_hrs, web_tuesday, 'g', label="tuesday")
plt.plot(time_hrs, web_wednesday, 'b', label="wednesday")
plt.title('Website Traffic')
plt.xlabel("Hours")
plt.ylabel("Number of costumers")
plt.show()
```
