```python
import matplotlib.pyplot as plt
from matplotlib import style

import numpy as np 
```

```python
rndm_num = np.random.rand(10)
rndm_num
```

```python
# style matplot
style.use('ggplot')

# plot random number
plt.plot(rndm_num, 'g', label='line one', linewidth=2)

# define x axis
plt.xlabel('Range')

# define y axis
plt.ylabel('Numbers')

# title
plt.title('1st-plot')

# showing plot
plt.legend()
plt.show()
```
