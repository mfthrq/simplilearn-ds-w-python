```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cause = 'Batuk', 'Tipes', 'Demam', 'Alzheimer', 'Infuenza', 'Kanker', 'Others'
percentile = [60,10,5,10,5,5,5]
```

```python
plt.figure(figsize=(10,10))
explode = (0.05,0,0,0,0,0,0)
plt.pie(percentile, labels=cause, explode=explode, autopct='%1.1f%%', startangle=70)
plt.axis('equal')
plt.title("Bogor - 2022 : Leading cause of death")
plt.show()
```
